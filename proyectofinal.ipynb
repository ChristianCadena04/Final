```python
datos= pd.read_csv("substancias_ilegales.csv")
```

```python
df = pd.DataFrame (datos)
```

```python
df.head()
```

```python
import numpy as pd
import time
plt.ticklabel_format(style='plain')
plt.figure(figsize=(30,10)) 
plt.savefig('Barras.png')

df.groupby('Cantón')['Cantidadkg'].sum().plot(kind='bar',legend='reverse',color={'yellow','blue','red','green'})
plt.ylabel("Cantones")
plt.xlabel("Cantidades")
plt.title("Cantidades Decomisadas por provincia : " + str(time.localtime().tm_year ) + "/"+ str(time.localtime().tm_mon,) + "/"+ str(time.localtime().tm_mday))
#Con este gráfico se puede visualizar el cantón donde existe la mayor cantidad de decomisos en el Ecuador
```

```python
import numpy as pd
import time
plt.ticklabel_format(style='plain')
plt.figure(figsize=(30,10)) 
plt.savefig('Barrasprovincias.png')

df.groupby('Provincia ')['Cantidadkg'].sum().plot(kind='bar',legend='reverse',color={'yellow','blue','red','green'})
plt.ylabel("Cantones")
plt.xlabel("Cantidades")
plt.title("Cantidades Decomisadas por provincia : " + str(time.localtime().tm_year ) + "/"+ str(time.localtime().tm_mon,) + "/"+ str(time.localtime().tm_mday))
#Con este gráfico se puede visualizar el cantón donde existe la mayor cantidad de decomisos en el Ecuador
```
